```python
#List comprehension
#new list =(output for)
l=[1,2,3,4,5]
res=[i*i for i in l]
print(res)
```

```python
#Lc
#new list=(output for if)
l1=[23,67,12,34,69]
res=[i*i for i in l1 if i>50]
print (res)
```

```python
#Lc
#new list=((st of if) if (cond) else (st of else) for)
l1=[23,67,12,34,69]
res=[i*i  if i>50 else 'x' for i in l1]
print (res)
```
