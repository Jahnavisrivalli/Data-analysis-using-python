```python
from bs4 import BeautifulSoup as bs
```

```python
import requests as rs
```

```python
url="https://aws.amazon.com/"
```

```python
page = rs.get(url)
```

```python
code = bs(page.text,'html')
print(code)
```

```python
t = code.find('title')
print(t)
```

EXTRACTING TABLE FROM A WEBPAGE

```python
url1 = "https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
page2 = rs.get(url1)
```

```python
code1 = bs(page2.text,'html')
print(code1)
```

```python
t1 = code1.find('table')
print(t1)
```

```python
h = t1.find_all('th')
```

```python
print(h)
```

```python
headings = [ i.text for i in h]
print(headings)
```

```python
import pandas as pd
```

```python
mydf = pd.DataFrame(columns = headings )        #command
print(mydf)
```

everything comes out from find is a list

```python
rows = t1.find_all('tr')
print(rows)
```

```python
tabledata = t1.find_all("td")
print(tabledata)
```

```python
tr = [i.text for i in  rows]
print(tr)
```

```python
td = [i.text for i in tabledata]
print(td)
print(td[0:4])
```

```python
nd = pd.DataFrame(columns =[ 'Name' ,' rank' , ' dollars' , 'company' ])
print(nd)
```

```python
start = 0
stop = 4
ind=0
for i in td:
  if stop<=40 :
    nd.loc[ind] = td[start : stop]
    start=start+4
    stop=stop+4
    ind=ind+1
print(nd)


```
