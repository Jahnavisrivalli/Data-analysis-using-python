```python
#There are 5 subjects,psychology,nano technology,medicine,anchology,anthrology.Two students have applied to do masters in any of these five streams with the line plot show  in which they can excel.The cut off for each subject is 70.
#s1=[67,45,73,71,57] ,  s2=[43,67,68,53,85]

x=['psychology','nano technology','medicine','anchology','anthrology']
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cutoff=[70,70,70,70,70]
plt.title("student marks")
plt.plot(x,s1,label='s1')
plt.plot(x,s2,label='s2')
plt.plot(x,cutoff,label='cutoff')
plt.legend()
```

```python
x=['psychology','nano technology','medicine','anchology','anthrology']
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cutoff=[70,70,70,70,70]
plt.title("student marks")
plt.scatter(x,s1,label='s1',color='b',marker="p")
plt.scatter(x,s2,label='s2',color='g',marker="H")
plt.plot(x,cutoff,label='cutoff',linestyle="--" , linewidth=5)
plt.legend()
```

```python
#An India vs Australia match is on live.India has given 298 as target for australia.The runs for every 5 overs is given as follows:[25,51,84,131,160,189,220,250,267,297].Australia have completed in 25 overs runs:[15,41,94,110,151,].plot overs vs runs for the both with blue is india ,yellow is aus.give required labels and lagend.the marker for india is flag and aus is plus.TAke overs from arange function.create arrays for every runs
import numpy as np
India=np.array([25,51,84,131,160,189,220,250,267,297])
Aus=np.array([15,41,94,110,151])
Indovers=np.arange(5,51,5)
Ausovers=np.arange(5,26,5)
plt.title("IND vs AUS")
plt.plot(India, Indovers,label='India',color='b',marker=">" , linewidth=2)

plt.plot(Aus , Ausovers,label='Aus',color='yellow',marker="+", linewidth=2)
plt.xlabel("overs")
plt.ylabel("runs")

plt.legend()
plt.grid()

```
