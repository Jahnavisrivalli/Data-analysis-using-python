```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz,plot_tree
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("/content/cancer patient data sets.csv")
df
```

```python
feature_col=['Smoking','Dry Cough','Genetic Risk','Snoring']
feature= df[feature_col]
Target=df.Level
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier()
```

```python
model.fit(feature,Target)
```

```python
model.predict([[3,3,3,4]])
```

```python
s=int(input("Enter the count of smoking:"))
c=int(input("Enter count of the dry cough:"))
r=int(input("Enter the count of the genetic risk:"))
n=int(input("Enter the count of snoring:"))
pred=model.predict([[s,c,r,n]])
if pred=='Low':
  print("No health issues")
elif pred=='Medium':
  print("average health cancer")
else:
  print("High health cancer")
```

```python

from sklearn.tree import export_graphviz,plot_tree
```

```python
plot_tree(model)
```

```python
RF=RandomForestClassifier(n_estimators=10)
```

```python
df1=pd.read_csv("/content/cancer patient data sets.csv")
df1
```

```python
feature_col=['Smoking','Dry Cough','Genetic Risk','Snoring']
feature= df1[feature_col]
Target=df1.Level
```

```python
RF.fit(feature,Target)
```

```python
model.predict([[3,3,3,4]])
```

```python
import os
output_dir = "tree visualisation"
os.makedirs(output_dir,exist_ok=True)
```

```python
RF.estimators_
```

```python
for i, tree in enumerate(RF.estimators_):
   tree_dot_file =  os.path.join(output_dir,f"tree_{i}.dot")
   tree_png_file = os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=["Smoking", "Dry Cough", "Genetic Risk","Snoring"],
                    class_names=[str(cls) for cls in RF.classes_], filled=True,rounded=True)
```
